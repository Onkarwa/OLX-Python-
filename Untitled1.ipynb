```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import seaborn as sns
import os
%matplotlib inline
```

```python
df=pd.read_csv('OLX_FINAL.csv')
```

```python
df.head()
```

```python
df['Brand'].value_counts().plot.bar()
plt.xlabel('Brands')
plt.title('Brands of Car')
```

```python
df['Year'].value_counts().plot.bar()
plt.xlabel('Years')
plt.title('Cars sales in year')



```

```python
var = df.groupby('State').Price.sum()
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.set_xlabel('States')
var.plot(kind='bar')
ax1.set_title("Car Sales in States")
```

```python
var = df.groupby('Model').Price.sum() 
fig = plt.figure(figsize=(11,5))
ax1 = fig.add_subplot(1,1,1)
ax1.set_xlabel('Model Of Car')
ax1.set_ylabel('Increase In price')
ax1.set_title("Model Vs Price")
var.plot(kind='line')
```

```python
var = df.groupby('Fuel').Price.sum()
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.set_xlabel('Fuel')
ax1.set_ylabel('Price')
ax1.set_title("Fuel Vs Price")
var.plot(kind='bar')
```
